```python
print("Choose the equation you want to solve")
print("1 - Quadratic Equation")
print("2 - Cubic Equation")
print("3 - Quartic Equation")

choice = int(input("Enter your choice: "))

# Quadratic equation 
if choice == 1:
    a = float(input("Enter value of a: "))
    b = float(input("Enter value of b: "))
    c = float(input("Enter value of c: "))

    d = (b*b) - (4*a*c) 

    root1 = (-b + d**0.5) / (2*a)
    root2 = (-b - d**0.5) / (2*a)

    print("Root 1 = ", root1)
    print("Root 2 = ", root2)

# Cubic equation 
elif choice == 2: 
    A = float(input("Enter A: "))
    B = float(input("Enter B: "))
    C = float(input("Enter C: "))
    D = float(input("Enter D: "))

    print("Your cubic equation is:")
    print(A, "x^3 +", B, "x^2 +", C, "x +", D, "= 0")

# Quartic equation 
elif choice == 3: 
    A = float(input("Enter A: "))
    B = float(input("Enter B: "))
    C = float(input("Enter C: "))
    D = float(input("Enter D: "))
    E = float(input("Enter E: "))

    print("Your quartic equation is:")
    print(A, "x^4 +", B, "x^3 +", C, "x^2 +", D, "x +", E, "= 0")

else: 
    print("Invalid choice")
    
    
    
```
