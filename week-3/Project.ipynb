```python
import tkinter as tk
import pandas as pd

# Function that runs when button is clicked 
def verify_employee():
    name = name_entry.get()
    dept = dept_entry.get()

    try: 
        # Load CSV file using pandas 
        data = pd.read_csv("C:/Users/aaa/OneDrive/Documents/GIG-logistics.csv")

        # Filter for matching employee
        match = data[
            (data["Name"].str.lower() == name.lower()) &
            (data["Department"].str.lower() == dept.lower())
            ]

        if not match.empty:
            # Get all members in same department 
            same_dept = data[data["Department"].str.lower() == dept.lower()]

            result = f"Welcome, {name}✨ from {dept}!\n\nOther members:\n"

            for member in same_dept["Name"]:
                if member.lower() != name.lower():
                    result += f"- {member}\n"

            output_label.config(text=result, fg="#6a0dad")
            
        else:
            output_label.config(text="Sorry, employee does not exist.")

    except Exception as e:
        output_label.config(text=f"Error: {e}", fg="red")
        
# Create main window 
window = tk.Tk()
window.title("Employee Verification System")
window.geometry("420x420")
window.configure(bg="#ffe6f0")

# Name input 
tk.Label(window, text="Name").pack()
name_entry = tk.Entry(window)
name_entry.pack()

# Department input 
tk.Label(window, text="Department").pack()
dept_entry = tk.Entry(window)
dept_entry.pack()

# Button 
tk.Button(window, text="Verify", command=verify_employee).pack()

# Output label 
output_label = tk.Label(window, text="")
output_label.pack()

# Run the GUI
window.mainloop()
            
```
